```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_ein = np.loadtxt("../csv-outputs/einstein-ring-output-2.csv", delimiter=",")
```

```python
#given csv with columns for r, theta, phi, x, y, and z
fig = plt.figure()
a = fig.add_subplot(111)

a.plot(data_ein[:,1], data_ein[:,2])

a.set_aspect(aspect=1)

title = "Einstein Ring in Schwarzchild Spacetime"
titleModifier = " Case 2"


plt.title(title+titleModifier)
plt.xlabel('x/M')
plt.ylabel('y/M')

circle1 = plt.Circle((0, 0), 2, color='k')
plt.gca().add_patch(circle1)

plt.savefig("../plots/einstein_ring_output"+"_"+titleModifier+".png", facecolor='w')
plt.show()
```
